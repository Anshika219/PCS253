```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
data={
    "trans_id":[1,2,3,4,5],
    "items":["milk,bread,eggs","butter,bread","eggs,milk","milk,bread","milk,butter"]
}
df=pd.DataFrame(data)
df
```

```python
#Encoding
```

```python
transactions=df['items'].apply(lambda x:x.split(','))
te=TransactionEncoder()
```

```python
te_data=te.fit(transactions).transform(transactions)
te_encoded=pd.DataFrame(te_data,columns=te.columns_)
```

```python
te_encoded.head()
```

```python
# Apriori Algorithm
```

```python
freq_itemsets=apriori(te_encoded,min_support=0.2,use_colnames=True)
```

```python
#Association Rule Mining
```

```python
rules=association_rules(freq_itemsets,metric="confidence",min_threshold=0.5)
```

```python
print("Association Rules")
print(rules[['antecedents','consequents','support','confidence','lift']])
```
