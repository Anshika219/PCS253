```python
#DFS
# Constructing a static graph with 6 nodes and 10 edges
graph = {
    'A': ['B', 'C', 'D'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F'],
    'D': ['A', 'B', 'E', 'F'],
    'E': ['B', 'D'],
    'F': ['C', 'D']
}


def dfs(graph, start, goal, visited=None):
    if visited is None:
        visited = set()
    print(start)
    if start==goal:
        return True

    visited.add(start)

    for neighbour in graph[start]:
        if neighbour not in visited:
            if dfs(graph, neighbour, goal, visited):
                return True
    return True
dfs(graph, 'A','E')
```

```python
#BFS
from collections import deque

graph = {
    'A': ['B', 'C', 'D'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F'],
    'D': ['A', 'B', 'E', 'F'],
    'E': ['B', 'D'],
    'F': ['C', 'D']
}

def bfs(graph, start,goal):
    visited = set()
    queue = deque([start])

    visited.add(start)
    path=[]

    while queue:
        node = queue.popleft()
        path.append(node)
        if node==goal:
            return path

        for neighbour in graph[node]:
            if neighbour not in visited:
                visited.add(neighbour)
                queue.append(neighbour)
    return path

bfs(graph, 'A','E')
```
