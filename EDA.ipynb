```python
import numpy as np
arr=np.array([[12,2,3],[2,3,6],[4,7,8]])
print(arr)
```

```python
print(arr.shape)
print(arr.ndim)
print(arr.size)
print(type(arr))
```

```python
print(arr.mean())
print(arr.std())
print(arr.var())
print(arr.max())
print(arr.min())
print(arr.sum())
```

```python
#linear algebra
a=np.array([[1,2,3],[2,4,6],[6,7,8]])
b=np.array([[1,3,5],[3,4,3],[7,8,9]])
print(a)
print(b)
```

```python
print(np.linalg.det(b))
print(np.linalg.matrix_rank(a))
print(np.linalg.inv(b))
print(np.linalg.eig(b))
```

```python
print(np.dot(a,b))
```

```python
print(np.trace(a))
```
