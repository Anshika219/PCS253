```python
#dfs
from collections import deque
def dfs(st,goal,graph,vis=None):
    if vis is None:
        vis = set()
    print(st);
    if st==goal:
        return True
    vis.add(st)
    for child in graph[st]:
        if child not in vis:
            if dfs(child,goal,graph,vis):
                return True
    return False
```

```python
graph={
    'A':['B','C'],
    'B':['A','D'],
    'C':['A','F'],
    'D':['B'],
    'E':['C','G'],
    'F':['C','E'],
    'G':['E']
}
```

```python
# START: A TO END: F
#OUTPUT: A B D C F
dfs('A','F',graph)
```
